```python
import tensorflow as tf
import pandas as pd
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def Load_FileName(path): # 해당 폴더에 있는 모든 파일 이름들 불러오는 코드
    folder_list = os.listdir(path)
    XData = []
    p = 0
    for i in folder_list:
        file_list = os.listdir(path + '/' + i)
        for j in file_list:
            print(path + '/' + i + '/' + file_list[p])
            c = cv2.imread(path + '/' + i + '/' + file_list[p])
            c = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
            print(c.shape)
            XData.append(c) # Data_set 리스트 안에 사진 정보를 차례대로 넣는다
            p += 1
        p = 0
    print(np.array(XData))
    return np.array(XData)# 리스트를 배열화 시킨다.

def Make_Y(path, Y_list):
    folder_name = os.listdir(path)
    Y = []
    p = 0
    for i in folder_name:
        file_name = os.listdir(path+'/'+ i)
        for j in file_name:
            Y.append(Y_list[p])
            print(p)
        p += 1
    return np.array(Y)
```

```python
X = Load_FileName('C:/Users/asdew32/Desktop/test/penguins')
# 테스트 데이터셋 가져오기

ValX = Load_FileName('C:/Users/asdew32/Desktop/test/validation')
# 검증을 위한 데이터셋

Y = Make_Y('C:/Users/asdew32/Desktop/test/penguins', ['A', 'B', 'C', 'D', 'E']) # 답지 데이터셋 만들기
ValY = Make_Y('C:/Users/asdew32/Desktop/test/validation', ['A', 'B', 'C', 'D', 'E']) # 검증을 위한 답지 데이터

# 왕관 펭귄 : 쇠푸른 펭귄 A, 왕관 펭귄 B, 황제 펭귄 C, 젠투 펭귄 D, 줄무늬 펭귄 E

Y = pd.get_dummies(Y) # 원 핫 인코딩
ValY = pd.get_dummies(ValY)
print(X.shape)
print(Y.shape)
print(ValX.shape)
print(ValY.shape)
```

```python
''' X_study = tf.keras.layers.Input(shape=[500,500,3])
H = tf.keras.layers.Conv2D(4, kernel_size=5, activation="relu")(X_study)
H = tf.keras.layers.MaxPooling2D()(H)
H = tf.keras.layers.Conv2D(8, kernel_size=3, activation="relu")(H)
H = tf.keras.layers.MaxPooling2D()(H)
H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(40, activation="relu")(H)
Y_study = tf.keras.layers.Dense(3, activation="softmax")(H)

model = tf.keras.models.Model(X_study, Y_study)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) '''

# 5개의 컨볼루션 3Dense층
studyX = tf.keras.layers.Input(shape=[227, 227, 3])

H = tf.keras.layers.Conv2D(96, kernel_size=11, strides=(4, 4), activation="relu")(studyX)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.MaxPool2D(pool_size=(3,3), strides=(2, 2))(H)

H = tf.keras.layers.Conv2D(256, kernel_size=5, padding="same", activation="relu")(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.MaxPool2D(pool_size=(3,3), strides=(2, 2))(H)

H = tf.keras.layers.Conv2D(384, kernel_size=3, padding="same", activation="relu")(H)

H = tf.keras.layers.Conv2D(384, kernel_size=3, padding="same", activation="relu")(H)

H = tf.keras.layers.Conv2D(256, kernel_size=3, padding="same", activation="relu")(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.MaxPool2D(pool_size=(3,3), strides=(2, 2))(H)

H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(4096, activation="relu")(H)
H = tf.keras.layers.Dense(4096, activation="relu")(H)
studyY = tf.keras.layers.Dense(5, activation="softmax")(H)

model = tf.keras.models.Model(studyX, studyY)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, Y, epochs = 30, validation_data = (ValX, ValY))
```
